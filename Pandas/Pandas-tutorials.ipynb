### convert data
* converters attr takes dict of key and value as function contains of the condition

```python
def convert_sales_cell(cell):
    if cell=="sales":
        return 'not sales'
    return cell

def convert_salary_cell(cell):
    if cell=="medium":
        return "m"
    return cell
    
df = pd.read_excel("new_excel.xlsx","Sheet1", converters= {
        'sales': convert_sales_cell,
        'salary': convert_salary_cell
    })
df
```

### write into excel

```python
df.to_excel("new_excel.xlsx", index=False, startrow=2, startcol=1)
```

### if you have two dataFrames and want to write them in two sheets

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
