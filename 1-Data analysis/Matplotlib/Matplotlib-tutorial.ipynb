### Bar Chart

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[50,40,70,10]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2,revenue,width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4, label="profit")
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.yticks(xpos,company)

plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit, label="profit")

plt.xlabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

### Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist(blood_sugar) 
# by default number of bins is set to 10
```

```python
plt.hist(blood_sugar, rwidth=0.8) 
```

```python
plt.hist([blood_sugar,blood_sugar_women], bins=[80,100,125,150], rwidth=0.8, color=['g','orange'], label=['men','women']) 
plt.title('Blood suger analysis')
plt.xlabel('suger range')
plt.ylabel('total number of patients')
plt.legend()
```

```python
plt.hist([blood_sugar,blood_sugar_women], bins=[80,100,125,150], rwidth=0.8, color=['g','orange'], label=['men','women'], orientation='horizontal') 
plt.title('Blood suger analysis')
plt.xlabel('suger range')
plt.ylabel('total number of patients')
plt.legend()
```
